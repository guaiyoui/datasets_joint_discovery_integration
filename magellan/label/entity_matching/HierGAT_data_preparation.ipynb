## Data Preparation

First convert 'train.csv', 'test.csv' and 'validate.csv' to 'txt' file, and do following operations.

```python
import pandas as pd
import os

datalake_folder = 'datalake'
dataset_file = 'train.txt'
output_file = 'train_sample_hiergat.txt'

def row_to_hiergat_str(row):
    s = ''
    for col, val in row.items():
        val = str(val).strip()
        s += f'COL {col} VAL {val} '
    return s.strip()

with open(dataset_file, 'r') as f_in, open(output_file, 'w') as f_out:
    header = next(f_in)  # skip header
    count = 0

    for line in f_in:
        if not line.strip():
            continue
        if count >= 2000:
            break

        try:
            ltable_name, l_id, rtable_name, r_id, label = line.strip().split(',')
            l_id, r_id = int(l_id), int(r_id)
            label = label.strip()

            l_path = os.path.join(datalake_folder, ltable_name)
            r_path = os.path.join(datalake_folder, rtable_name)
            l_df = pd.read_csv(l_path)
            r_df = pd.read_csv(r_path)

            # set first column to the key column
            l_key = l_df.columns[0]
            r_key = r_df.columns[0]

            # Find the row corresponding to the primary key value
            l_row = l_df[l_df[l_key] == l_id].iloc[0]
            r_row = r_df[r_df[r_key] == r_id].iloc[0]

            # convert to 'train_sample_csv'
            l_str = row_to_hiergat_str(l_row)
            r_str = row_to_hiergat_str(r_row)
            f_out.write(f'{l_str}\t{r_str}\t{label}\n')

            count += 1

        except Exception as e:
            print(f'Error processing line {count+1}: {line.strip()}')
            print(f'Exception: {e}')
```
